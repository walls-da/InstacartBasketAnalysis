### 4. Histograms & Scatterplots

```python
# Create a histogram

df_op['prices'].plot.hist(bins = 25)
```

##### Observations:
1. Only one bar is present. Data check on 'prices' is necessary

```python
# Data check on prices column

df_op['prices'].describe()
```

```python
# Create a scatterplot for data distribution to see any outliers

sns.scatterplot(x = 'prices', y = 'prices',data = df_op)
```

```python
# Check if price is above a "normal" threshold

df_op.loc[df_op['prices'] > 100]
```

##### Observations: 
1. 5,144 rows of outliers. Two price values that stick out: 14,900 and 99,999. Consult clients and stakeholders to confirm.

```python
# Mark outliers as "missing" 

df_op.loc[df_op['prices'] >100, 'prices'] = np.nan
```

```python
df_op['prices'].max()
```

```python
# Retry of histogram

df_op['prices'].plot.hist(bins = 25)
```

```python
# Create a more distinguishable histogram

df_op['prices'].plot.hist(bins = 70)
```

##### Observations:
1. Most products are between 1 and 15 dollars
2. A few are higher priced at 15 to 25 dollars

```python
# Saving histogram into Jupyter & exporting as an image

hist = df_op['prices'].plot.hist(bins = 70)
hist.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'hist_prices.png'))
```
